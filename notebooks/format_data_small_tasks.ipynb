```python
import numpy as np
from datetime import datetime
import pickle
```

# Format operations

```python
def date2array(s):
    # Define the format of the date string
    date_format = "%B %d, %Y."
    # Parse the date string into a datetime object
    parsed_date = datetime.strptime(s, date_format)
    # Format the date into a different string format
    formatted_date = [parsed_date.year, parsed_date.month, parsed_date.day]
    return formatted_date
```

```python
def generate_operation(i, j, k):
    symbol = ["+", "-", "*", "/"]
    string = "{} {} {}".format(i, symbol[k], j)
    return string

def gather_all_operations(n=10):
    result_list=[]
    for i in range(n):
        for j in range(n):
            for k in range(3):
                result_list.append([generate_operation(i, j, k), [i, j, k], eval(f'{generate_operation(i, j, k)}')])
    return result_list

operation_list = gather_all_operations(10)
print(len(operation_list))
```

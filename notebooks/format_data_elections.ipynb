```python
import numpy as np
from datetime import datetime
import pickle
from dateutil import parser
```

# Format Election 2021, 2022, 2024 list

```python
def date2array(s):
    # Define the format of the date string
    date_format = "%B %d, %Y."
    # Parse the date string into a datetime object
    parsed_date = parser.parse(s)#, date_format)
    print(parsed_date)
    # Format the date into a different string format
    formatted_date = parsed_date#[parsed_date.year, parsed_date.month, parsed_date.day]
    return formatted_date
```

```python
word_list = []
with open("/srv/public/Thibaud/datasets/LLMs/MRI/texts/elections_2010", "r") as f:
    for w in f:
        if len(w)>1:
            w = w.split("\n")[0]
            w = w.split(",")
            print(w)
            # Sanity check
            if not len(w)==3:
                print(w)
                raise "Not good formating--.dsdgs"
            else :
                word_list.append([w[0], date2array(w[1]), w[2]])

print(word_list)
```

```python
to_del = []
print(len(word_list))
print(len(set(word_list)))
for i, w in enumerate(word_list):
    for ww in word_list[i+1:]:
        if w == ww :
            to_del.append(i)
            print(w, ww)
for i in range(len(to_del), 0, -1):
    print(to_del[i-1])
    del word_list[to_del[i-1]]
print(len(word_list))
```

```python
with open("/srv/public/Thibaud/datasets/LLMs/MRI/pkls/elections2010.pkl", "wb") as fp:
    pickle.dump(word_list, fp)
    fp.close()
```

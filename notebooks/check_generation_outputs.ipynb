## Symple script to check LLM outputs

This script prints out all the original outputs from the LLM _Llama3-8B-Instruct_ on the prompts of our dataset.

The original activation files are not in the git because of their size.

```python
import pickle
import sys
import pandas as pd

sys.path.append('../src')
from projection import *
```

```python
def print_input_output(name, label):
    # Load the data from the pickle file
    file = "../data/projected/{}.pkl".format(name)
    with open(file, 'rb') as f:
        data = pickle.load(f)

    # Filter desired elements
    label_df = pd.DataFrame([elmt.__dict__ for elmt in data])
    label_df = label_df[label_df['label'] == label]
    print("<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<")
    print(">>> Inputs and Outputs for Llama3-8B-Instruct on {} {} entities >>>>>".format(name, ["Known", "Unkmown"][label]))
    print("<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<")
    for i in range(len(label_df)):
        print("intput:", label_df.iloc[i]["input_text"])
        print("output:", label_df.iloc[i]["output_text"])
        print()
```
